## Parte A вҖ” AnГЎlisis de Datos con Pandas

Analiza el siguiente dataset de ventas.

```python
import pandas as pd
import numpy as np

# Dataset de ventas вҖ” NO MODIFICAR
np.random.seed(42)
datos_ventas = {
    "producto": ["Laptop", "Mouse", "Teclado", "Monitor", "Laptop",
                 "Mouse", "Teclado", "Monitor", "Laptop", "Mouse",
                 "Teclado", "Monitor", "Laptop", "Mouse", "Teclado",
                 "Monitor", "Laptop", "Mouse", "Teclado", "Monitor"],
    "categoria": ["ElectrГіnica", "Accesorios", "Accesorios", "ElectrГіnica", "ElectrГіnica",
                   "Accesorios", "Accesorios", "ElectrГіnica", "ElectrГіnica", "Accesorios",
                   "Accesorios", "ElectrГіnica", "ElectrГіnica", "Accesorios", "Accesorios",
                   "ElectrГіnica", "ElectrГіnica", "Accesorios", "Accesorios", "ElectrГіnica"],
    "precio": [999, 25, 75, 350, 1099, 30, 80, 400, 899, 20,
               65, 320, 1199, 35, 90, 380, 950, 28, 70, 360],
    "cantidad": [3, 15, 8, 5, 2, 20, 10, 3, 4, 25,
                 12, 4, 1, 18, 7, 6, 3, 22, 9, 5],
    "mes": ["Ene", "Ene", "Ene", "Ene", "Feb", "Feb", "Feb", "Feb",
            "Mar", "Mar", "Mar", "Mar", "Abr", "Abr", "Abr", "Abr",
            "May", "May", "May", "May"]
}

df = pd.DataFrame(datos_ventas)
df["ingreso_total"] = df["precio"] * df["cantidad"]
print(df.head(20))
print(f"Shape: {df.shape}")
```

### A.1 вҖ” Ingreso total global

Calcula la suma de la columna `ingreso_total`.

```python
ingreso_total_global = df['ingreso_total'].sum() # Tu cГЎlculo aquГӯ
print(f"Ingreso total: {ingreso_total_global}")
```

### A.2 вҖ” Producto con mayor ingreso total

ВҝQuГ© producto genera mГЎs ingreso total (sumando todas sus ventas)?

```python
producto_ingreso_total = df.groupby('producto')['ingreso_total'].sum()
print("Ingreso total por producto:")
print(producto_ingreso_total)

producto_top = producto_ingreso_total.idxmax()
print(f"\nEl producto con mayor ingreso total es: {producto_top}")
```

```python
# A.2 вҖ” Producto con mayor ingreso
producto_top = df.groupby('producto')['ingreso_total'].sum().idxmax()  # Tu cГЎlculo aquГӯ (string)
print(f"Producto top: {producto_top}")
```

```python
# ValidaciГіn A.2 вҖ” NO MODIFICAR
assert producto_top == "Laptop", f"Error: deberГӯa ser 'Laptop', obtuviste '{producto_top}'"
print("вң… A.2 вҖ” Producto top: CORRECTO")
```

### A.3 вҖ” Media de precio por categorГӯa

Calcula el precio medio por categorГӯa. Devuelve un diccionario redondeado a 2 decimales.

```python
# A.3 вҖ” Precio medio por categorГӯa
precio_medio_categoria = df.groupby('categoria')['precio'].mean().round(2).to_dict()  # Tu cГЎlculo aquГӯ (dict)
print(f"Precio medio: {precio_medio_categoria}")
```

### A.4 вҖ” Mes mГЎs rentable

## Parte B вҖ” Red Neuronal con Keras

Construye un clasificador para **Fashion-MNIST** y alcanza un accuracy > 85%.

```python
import tensorflow as tf
from tensorflow import keras

# Cargar Fashion-MNIST
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

# Normalizar
x_train = x_train / 255.0
x_test = x_test / 255.0

# Nombres de las clases
clases = ["Camiseta", "PantalГіn", "Jersey", "Vestido", "Abrigo",
          "Sandalia", "Camisa", "Zapatilla", "Bolso", "Bota"]

print(f"Train: {x_train.shape}, Test: {x_test.shape}")
print(f"Clases: {clases}")
```

### B.1 вҖ” Construye el modelo

Crea un modelo `Sequential` con:
- Capa `Flatten` de input
- Al menos 2 capas `Dense` con activaciГіn `relu`
- Capa de salida con 10 neuronas y `softmax`

> рҹ’Ў Pide a Gemini: *"Crea un modelo Keras Sequential para clasificar Fashion-MNIST con al menos 2 capas ocultas"*

```python
# B.1 вҖ” Define tu modelo aquГӯ
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)), # Capa de input Flatten
    keras.layers.Dense(128, activation='relu'), # Primera capa oculta Dense con relu
    keras.layers.Dense(64, activation='relu'),  # Segunda capa oculta Dense con relu
    keras.layers.Dense(10, activation='softmax') # Capa de salida con 10 neuronas y softmax
])

# Compilar
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model.summary()
```

### B.2 вҖ” Entrena el modelo

Entrena el modelo y consigue un **accuracy en test > 85%**.

```python
# B.2 вҖ” Entrena tu modelo
history = model.fit(x_train, y_train, epochs=30, batch_size=32, validation_split=0.1)
```

```python
# B.2 вҖ” EvalГәa en test
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
# accuracy = None  # Reemplaza con el accuracy obtenido (ej: 0.88)
print(f"Accuracy en test: {accuracy}")
```

```python
# ValidaciГіn B.2 вҖ” NO MODIFICAR
assert accuracy is not None, "Error: entrena el modelo y asigna el accuracy"
assert accuracy > 0.85, f"Error: accuracy debe ser > 0.85, obtuviste {accuracy:.4f}"
print(f"вң… B.2 вҖ” Accuracy: {accuracy:.4f} (> 85%): CORRECTO")
```

### B.3 вҖ” PredicciГіn

Predice las primeras 5 imГЎgenes del test set y muГ©stralas junto a su predicciГіn.

```python
# B.3 вҖ” Predice y visualiza las primeras 5 imГЎgenes
import matplotlib.pyplot as plt
import numpy as np

# Tu cГіdigo aquГӯ вҖ” muestra 5 imГЎgenes con su predicciГіn
# Pide a Gemini: "Muestra las primeras 5 imГЎgenes del test set con la predicciГіn del modelo"

# Realizar predicciones para las primeras 5 imГЎgenes del conjunto de prueba
predictions = model.predict(x_test[:5])
predicted_classes = np.argmax(predictions, axis=1)

plt.figure(figsize=(10, 5))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(x_test[i], cmap='gray')
    plt.title(f"Real: {clases[y_test[i]]}\nPred: {clases[predicted_classes[i]]}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```

## Parte C вҖ” Escribe una Spec (Opcional)

Si quieres practicar, escribe una **especificaciГіn en lenguaje natural** para una herramienta Гәtil (como vimos en clase). Tu spec debe incluir: descripciГіn, inputs, outputs, requisitos tГ©cnicos y ejemplo de uso. DespuГ©s, pГ©gala en Gemini y copia el cГіdigo generado en la celda siguiente. Esto no es obligatorio pero es excelente prГЎctica de prompting.

```python
# C.1 вҖ” Tu especificaciГіn
spec = """
# Spec: Calculadora de Presupuesto de Eventos Corporativos

## DescripciГіn
Una herramienta donde un empleado ingresa los detalles de un evento (una cena de equipo, un workshop o una fiesta de fin de aГұo) y la API valida si el presupuesto estГЎ dentro de las polГӯticas de la empresa.

## Inputs
Nombre del evento: event_name (string)
NГәmero de asistentes: attendees (integer)
Tipo de evento: event_type (CategorГӯas: 'Workshop', 'TeamBuilding', 'Dinner')
Presupuesto estimado: total_estimated_cost (float)

## Reglas de negocio:
Calcula el cost_per_person (total_estimated_cost/attendees)
Si el cost_per_person es mayor a 100, el estado es 'Rejected: Over Budget'.
Si el cost_per_person estГЎ entre 50 y 100, el estado es 'Needs Manager Approval'.
Si es menor a 50, el estado es 'Auto-Approved'.

## Outputs
Status de la aprobaciГіn segГәn polГӯtica

## Requisitos tГ©cnicos
Devuelve un JSON con el resumen, el costo por persona y el estado. Usa Pydantic para validar los datos y asegГәrate de que la documentaciГіn de Swagger sea clara."

## Ejemplo
event_name = "Workshop de Python"
event_type = "Workshop"
attendees = 3
total_estimated_cost = 100
status: "Auto-Approved"

"""
print(spec)
```

```python
# C.2 вҖ” CГіdigo generado por Gemini a partir de tu spec
# Pega aquГӯ el resultado
from fastapi import FastAPI
from pydantic import BaseModel
from enum import Enum

app = FastAPI(title="EventPlanner Spec")

class EventType(str, Enum):
    workshop = "Workshop"
    teambuilding = "TeamBuilding"
    dinner = "Dinner"

class Event(BaseModel):
    event_name: str
    event_type: EventType
    attendees: int
    total_estimated_cost: float

@app.post("/analyze-event")
def analyze_event(event: Event):
    cost_per_person = event.total_estimated_cost / event.attendees

    if cost_per_person > 100:
        status = "Rejected: Over Budget"
    elif 50 <= cost_per_person <= 100:
        status = "Needs Manager Approval"
    else:
        status = "Auto-Approved"

    return {
        "event": event.event_name,
        "cost_per_person": round(cost_per_person, 2),
        "status": status,
        "policy_applied": "Global Expense Policy 2024"
    }
# link de la Spec: https://ejercicio2-app-eventos.onrender.com/docs
```
